```python
import datetime

now = datetime.datetime.now()
month = now.month

if 3 <= month <= 5:
    print('현재는 봄입니다.')
elif 6 <= month <= 8:
    print('현재는 여름입니다.')
elif 9 <= month <= 11:
    print('현재는 가을입니다.')
else:
    print('현재는 겨울입니다.')
```

```python
import datetime
now = datetime.datetime.now()

hour = now.hour
if 2 <= hour <= 6:
    print('새벽입니다.')
elif 7 <= hour <= 11:
    print('오전입니다')
elif 12 <= hour <= 20:
    print('오후입니다')
else:
    print('밤입니다.')
```

```python
score = float(input("학점 입력"))

if score == 4.5:
    print('신')
elif 4.2 <= score < 4.5:
    print('교수님의 사랑')
elif 3.5 <= score < 4.2:
    print('현 체제의 수호자')
elif 2.8 <= score < 3.5:
    print('일반인')
elif 2.3 <= score < 2.8:
    print('일탈을 꿈꾸는 소시민')
elif 1.75<= score < 2.3:
    print('오락문화의 선구자')
elif 1.0 <= score < 1.75:
    print('불가촉천민')
elif 0.5 <= score < 1.0:
    print('자벌레')
elif 0 <= score < 0.5:
    print('플랑크톤')
elif score == 0:
    print('시대를 앞서가는 혁명의 씨앗')
```
